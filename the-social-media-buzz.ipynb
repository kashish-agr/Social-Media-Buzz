## Data Cleaning


> Data cleaning is a common and very import task when working with data. 

## Importing All Esstential Libraries 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
from IPython.display import display
```

### Uploading original datasets


```python
content = pd.read_csv('/kaggle/input/forage-accenture-virtual-intern/Content.csv', index_col=0)
reaction = pd.read_csv('/kaggle/input/forage-accenture-virtual-intern/Reactions.csv', index_col=0)
reaction_type = pd.read_csv('/kaggle/input/forage-accenture-virtual-intern/ReactionTypes.csv', index_col=0)
```

```python
display(content)
display(reaction)
display(reaction_type)
```

###  Insights of content dataset :
* URL and User ID are not relevent to this task.So, remove it .
* Rename the 'Type' column to 'Content_type'.
* Check the unique values of type and category to identify duplication and misspelled values
* Remove all nan values 

```python
content.info()
```

```python
content.rename(columns={'Type':'Content_type'}, inplace=True)
clear_content = content.drop(columns=['User ID','URL'])
display(clear_content)
```

#### Let's print the  unique values of category and content-type to identify duplicates/misspellings

```python
print(clear_content['Content_type'].unique())
```

```python
clear_content['Category'].unique()
```

#### Replace all Duplicates and misspelled classes

```python
replacements = {
    '"': '',
    'Studying': 'studying',
    'Soccer': 'soccer',
    'Animals': 'animals',
    'Veganism': 'veganism',
    'Healthy Eating': 'healthy eating',
    'Fitness': 'fitness',
    'Travel': 'travel',
    'Education': 'education',
    'Technology': 'technology',
    'Culture': 'culture',
    'Food': 'food',
    'Science': 'science',
    'Public Speaking': 'public speaking'
}
clear_content['Category'] = clear_content['Category'].replace(replacements, regex=True)
```

```python
clear_content['Category'].unique()
```

#### Remove all null values 

```python
clear_content.dropna(how ='any', inplace = True)
```

```python
display(clear_content)
```

### Insights for Reaction dataset :
* Convert object datatype of 'Datetime' to datatime datatype
* Remove irrelevent feature 'User ID'
* Rename 'Type' to 'Reaction-type'.
* Check for unique values of type to identify duplicates or misspelled values.
* Remove all nan values 

```python
reaction.info()
```

#### Convert object to datatime datatype

```python
reaction['Datetime'] = pd.to_datetime(reaction['Datetime'])
```

#### Drop Unusal Column 

```python
clear_reactions = reaction.drop(columns='User ID')
```

#### Drop null values 

```python
clear_reactions.dropna(how='any', inplace = True)
```

```python
clear_reactions.rename(columns={'Type':'Reaction-type'}, inplace= True)
```

```python
clear_reactions.info()
```

```python
clear_reactions['Reaction-type'].unique()
```

```python
display(clear_reactions)
```

### Insights of Reaction_type:
* Rename 'Type' to 'Reaction-type'.
* Check for unique values of type to identify duplicates or misspelled values.
* Remove all nan values 

```python
reaction_type.info()
```

```python
reaction_type.rename(columns={'Type':'Reaction-type'}, inplace= True)
```

```python
reaction_type['Reaction-type'].unique()
```

```python
reaction_type['Sentiment'].unique()
```

```python
clean_reaction_type = reaction_type.dropna(how ='any')
```

```python
display(clean_reaction_type)
```

## Merging the Datasets 

```python
cleaned_dataset = pd.merge(clear_content , clear_reactions , how = 'right', on ='Content ID')
cleaned_dataset = pd.merge(cleaned_dataset , clean_reaction_type , how = 'left', on ='Reaction-type')
cleaned_dataset.dropna(how = 'any' , inplace =True)
display(cleaned_dataset)
```

## Find Top 5 categories based on highest score

```python
top_5_categories = cleaned_dataset.groupby(['Category'])['Score'].sum().sort_values(ascending = False).head(5)
```

```python
top_5category = top_5_categories.reset_index()
top_5category.column=['Category','Popularity ']
display(top_5category)
```
